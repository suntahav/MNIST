```python
X = train_data.iloc[:,1:].values
y = train_data.iloc[:,0].values
X.shape
```

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=17)
X_train = X_train/255
X_test  = X_test/255
```

```python
import keras
from keras.models import Sequential
from keras.layers.core import Dense,Dropout
from keras.utils import np_utils
from keras.layers.advanced_activations import LeakyReLU
from keras import regularizers
```

```python
#one hot encoding
total_classes = 10
y_train = np_utils.to_categorical(y_train,num_classes=total_classes)
y_test  = np_utils.to_categorical(y_test,num_classes=total_classes)

y_train[0]
```

```python
#model building
classifier = Sequential()

# classifier.add(Dense(512,activation='linear',input_shape=(784,)))
classifier.add(Dense(512,input_shape=(784,),kernel_regularizer=regularizers.l2(0.0001)))
classifier.add(LeakyReLU(alpha=0.001))
classifier.add(Dropout(0.17))
# classifier.add(Dense(392,activation='relu'))
classifier.add(Dense(392,kernel_regularizer=regularizers.l2(0.0001)))
classifier.add(LeakyReLU(alpha=0.001))
classifier.add(Dropout(0.20))

classifier.add(Dense(256,kernel_regularizer=regularizers.l2(0.0001)))
classifier.add(LeakyReLU(alpha=0.001))
classifier.add(Dropout(0.20))

classifier.add(Dense(10,activation='softmax'))
```

```python
classifier.compile(optimizer='adam',loss = 'categorical_crossentropy',metrics=['accuracy'])
```

```python
result = classifier.fit(X_train,y_train,batch_size=32,epochs=30,verbose=2,validation_data=(X_test,y_test))
```
